```r
library(MPV)
data<-data(table.b2)
y.lm <- lm(y ~ x1 + x2 + x3 + x4 + x5, data=table.b2)
summary(y.lm)
```

```r
##### NORMAL PROBABILITY PLOT OF RESIDUALS #####
# resid(y.lm);
# plot(density(resid(y.lm)))
y.res = resid(y.lm)
qqnorm(y.res, 
      ylab="Residuals", xlab="Normal Scores", main="Normal Probability plot of Residuals")
qqline(y.res)


```

```r
##### NORMAL PROBABILITY PLOT OF Studentized RESIDUALS #####

y.stdres = rstandard(y.lm)
qqnorm(y.stdres, 
      ylab="Residuals", xlab="Normal Scores", main="Normal Probability plot of studentized Residuals")
qqline(y.stdres)
```

```r
#### Residuals vs Predicted Response ####
y.predicted = predict(y.lm)
plot(y.predicted, y.res, main="Residuals vs Predicted Response")
abline(0,0)
```

```r
#### Studentized Residuals vs Predicted Response ####
y.predicted = predict(y.lm)
plot(y.predicted, y.stdres, main="Studentized Residuals vs Predicted Response")
abline(0,0)
```
